```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(
    r"C:\Users\olham\Desktop\New folder\student-por.csv",
    sep=";"
)

df.head()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x='sex', y='G3', data=df)

plt.xlabel("Gender")
plt.ylabel("Final Grade (G3)")
plt.title("Grade Distribution by Gender")

plt.show()
```

```python
avg = df.groupby('sex')[['G1','G2','G3']].mean().reset_index()

avg_melted = avg.melt(id_vars='sex', 
                      value_vars=['G1','G2','G3'],
                      var_name='Exam',
                      value_name='Average Grade')
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x='sex', y='Average Grade', 
             hue='Exam', 
             data=avg_melted,
             marker='o')

plt.title("Average Grades by Gender")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='sex', y='G3', data=df)

plt.title("Average Final Grade by Gender")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='G3', hue='sex', kde=True)

plt.title("Distribution of Final Grades")
plt.show()
```
