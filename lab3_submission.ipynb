```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df['G3'], bins=20)
plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['studytime'], bins=4)
plt.title("Distribution of Study Time")
plt.xlabel("Study Time Level")
plt.ylabel("Count")
plt.show()
```

```python
# =========================
# showing the avg by gender 
# =========================
avg_g3_by_sex = df.groupby("sex")["G3"].mean()
avg_g3_by_sex.plot(kind="bar")
plt.ylabel("Average G3")
plt.title("Average Final Grade (G3) by Sex")
plt.show()
```

```python
df.duplicated().sum()
```

```python
sns.scatterplot(x='studytime', y='G3', data=df)
plt.title('Study Time vs Final Grade')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Full Correlation Matrix")
plt.show()
```

```python
grade_trend = df[['G1', 'G2', 'G3']].mean()
grade_trend
```

```python
plt.figure(figsize=(8,5))
grade_trend.plot(marker='o')
plt.title("Average Grade Progression (G1 → G2 → G3)")
plt.ylabel("Average Grade")
plt.xlabel("Exam Period")
plt.show()
```

```python
studytime_performance = df.groupby('studytime')['G3'].mean().sort_values(ascending=False)

plt.figure(figsize=(6,4))
studytime_performance.plot(kind='bar')
plt.title("Average Final Grade by Study Time")
plt.ylabel("Average G3")
plt.show()

studytime_performance
```

```python
address_avg = df.groupby('address')['G3'].mean().sort_values(ascending=False)

plt.figure(figsize=(6,4))
address_avg.plot(kind='bar')
plt.title("Average Final Grade by Address")
plt.ylabel("Average G3")
plt.show()

address_avg
```
