```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (change path if needed)
df = pd.read_csv(
    r"C:\Users\olham\Desktop\New folder\student-por.csv",
    sep=";"
)

# Calculate average final grade (G3) by gender
avg_grades = df.groupby('sex')['G3'].mean()

# Prepare x and y
x = avg_grades.index          # ['F', 'M']
y = avg_grades.values         # average grades

# Plot (same style as your example)
plt.figure(figsize=(5,5))
plt.plot(x, y, 'r')           # red line

plt.xlabel('Gender')
plt.ylabel('Average Final Grade (G3)')
plt.title('Average Grade Difference Between Girls and Boys')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(
    r"C:\Users\olham\Desktop\New folder\student-por.csv",
    sep=";"
)

# Calculate averages
avg_G3 = df.groupby('sex')['G3'].mean()
avg_G1 = df.groupby('sex')['G1'].mean()

x = avg_G3.index   # F , M

# Create blank canvas
fig = plt.figure()

# Main axes
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Insert axes (small plot inside)
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

# ---- Big Plot (G3) ----
axes1.plot(x, avg_G3.values, 'b')
axes1.set_xlabel('Gender')
axes1.set_ylabel('Average G3')
axes1.set_title('Final Grade (G3)')

# ---- Small Plot (G1) ----
axes2.plot(x, avg_G1.values, 'r')
axes2.set_xlabel('Gender')
axes2.set_ylabel('Average G1')
axes2.set_title('First Grade (G1)')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(
    r"C:\Users\olham\Desktop\New folder\student-por.csv",
    sep=";"
)

# Calculate averages
avg_G1 = df.groupby('sex')['G1'].mean()
avg_G3 = df.groupby('sex')['G3'].mean()

x = avg_G1.index   # F , M

# Create figure and axes
fig, ax = plt.subplots()

# Plot both lines
ax.plot(x, avg_G1.values, label='G1 (First Grade)')
ax.plot(x, avg_G3.values, label='G3 (Final Grade)')

# Labels & title
ax.set_xlabel('Gender')
ax.set_ylabel('Average Grade')
ax.set_title('Comparison of G1 and G3 by Gender')

# Legend
ax.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(
    r"C:\Users\olham\Desktop\New folder\student-por.csv",
    sep=";"
)

# Calculate averages
avg = df.groupby('sex')[['G1','G2','G3']].mean()

x = avg.index   # F , M

fig, ax = plt.subplots(figsize=(8,5))

# G1 - blue dashed
ax.plot(x, avg['G1'], color='blue', lw=2, ls='--', marker='o',
        markersize=8, label='G1')

# G2 - green dotted
ax.plot(x, avg['G2'], color='green', lw=2, ls=':', marker='s',
        markersize=8, label='G2')

# G3 - red solid
ax.plot(x, avg['G3'], color='red', lw=3, ls='-', marker='D',
        markersize=9, label='G3')

ax.set_xlabel('Gender')
ax.set_ylabel('Average Grade')
ax.set_title('G1, G2, G3 Comparison by Gender')

ax.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(
    r"C:\Users\olham\Desktop\New folder\student-por.csv",
    sep=";"
)

# Separate grades by gender
girls = df[df['sex'] == 'F']['G3']
boys = df[df['sex'] == 'M']['G3']

# Create boxplot
plt.figure(figsize=(6,5))
plt.boxplot([girls, boys], vert=True, patch_artist=True)

plt.xticks([1,2], ['Girls', 'Boys'])
plt.ylabel('Final Grade (G3)')
plt.title('Grade Distribution by Gender')

plt.show()
```
